```python
# Descrição:
# Dados extraídos da Polícia sobre individuos presos em Toronto - Canadá, por posse de maconha. Os dados são uma amostra de um grande dataset, usado em uma série de artigos no jornal 'Toronto Star'.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o conjunto de dados
data = pd.read_csv('Arrests.csv')
```

```python
# Ocorrências por ano

# Contar o número de ocorrências para cada ano
year_counts = data['year'].value_counts().sort_index()

# Criar o gráfico de linha
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-')

# Definir os rótulos e títulos
plt.xlabel('Ano')
plt.ylabel('Número de Prisões')
plt.title('Prisões por Posse de Maconha ao Longo do Tempo')

# Exibir o gráfico de linha
plt.show()
```

```python
# Ocorrências por idade

# Contar o número de ocorrências para cada idade
age_counts = data['age'].value_counts().sort_index()

# Criar o gráfico de barras
plt.bar(age_counts.index, age_counts.values)

# Definir os rótulos e títulos
plt.xlabel('Idades')
plt.ylabel('Número de Ocorrências')
plt.title('Distribuição das Idades')

# Exibir o gráfico de barras
plt.show()
```

```python
# Ocorrências por Cor e Sexo

# Calcular a contagem de ocorrências para cada cor e sexo
color_counts = data['colour'].value_counts()
sex_counts = data['sex'].value_counts()

# Criar uma figura com subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Gráfico de pizza para a porcentagem de ocorrências por cor
ax1.pie(color_counts.values, labels=color_counts.index, autopct='%1.1f%%')
ax1.set_title('Porcentagem de Ocorrências por Cor')

# Gráfico de pizza para a porcentagem de ocorrências por sexo
ax2.pie(sex_counts.values, labels=sex_counts.index, autopct='%1.1f%%')
ax2.set_title('Porcentagem de Ocorrências por Sexo')

# Ajustar o layout da figura
plt.tight_layout()

# Exibir o gráfico de pizza
plt.show()
```
